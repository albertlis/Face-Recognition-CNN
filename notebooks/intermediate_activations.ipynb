```python
from keras.models import load_model

model = load_model('../models/standard_model_preprocessed.h5')
# model.summary()
```

```python
from keras.preprocessing import image

img = image.load_img('../data/histogram_stretched/train/1/3292.png', target_size=(96, 96),
                     color_mode='grayscale')
img_tensor = image.img_to_array(img)
img_tensor /= 255
```

```python
import matplotlib.pyplot as plt

plt.imshow(img_tensor, cmap='gray')
plt.show()
```

```python
from keras_conv_visualizer.intermediate_activations import IntermediateActivations

int_activations = IntermediateActivations(model)
int_activations.plot_intermediate_activations(img_tensor)
```
